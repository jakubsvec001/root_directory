```python
import pandas as pd
from urllib.parse import unquote
```

```python
data = pd.read_csv('seed_categories/cat_pages_mathematics_d6.csv', sep=',', encoding='utf-8')
```

```python
data.head(30)
```

```python
def clean(row):
    base = 'https://en.wikipedia.org/wiki/'
    base_len = len(base)
    return row[base_len:]
```

```python
data_stripped = data.iloc[:, 0].apply(clean)
```

```python
def yield_20(column):
    for i in range(len(column)):
        print(i)
        yield column[(i*100):(1+i)*100]
    
```

```python
data_stripped[13]
```

```python
twenty = yield_20(data_stripped)
```

```python
next(twenty)
```

```python
ascii_list = []
for row in data_stripped:
    if '%' in row:
        ascii_list.append(row)
```

```python
import csv
```

```python
with open('ascii_test.csv', 'w') as fout:
    writer = csv.writer(fout)
    for string in ascii_list:
        string = unquote(string)
        writer.writerow([string])
```

```python
ascii_out = []
with open('ascii_test.csv', 'r') as fin:
    reader = csv.reader(fin)
    for row in reader:
        ascii_out.append(row[0])
```

```python
full = []
for row in data_stripped:
    full.append(unquote(row))
```

```python
iterated = yield_20(full)
```
